```python
from textblob import TextBlob

# Metin dosyasını yükleyin
file_path = '/content/Hikayem.txt'

# Dosyayı aç ve oku
with open(file_path, 'r', encoding='utf-8') as file:
    text = file.read()

# Cümleleri ayır
blob = TextBlob(text)
sentences = blob.sentences

# Olumlu ve olumsuz cümle sayısını hesapla
positive_count = 0
negative_count = 0

for sentence in sentences:
    if sentence.sentiment.polarity > 0:
        positive_count += 1
    elif sentence.sentiment.polarity < 0:
        negative_count += 1

print("Olumlu cümle sayısı:", positive_count)
print("Olumsuz cümle sayısı:", negative_count)
```

```python
def textblob_sentiment_analysis(text):
    blob = TextBlob(text)
    polarity = blob.sentiment.polarity
    if polarity > 0:
        return "Olumlu"
    elif polarity < 0:
        return "Olumsuz"
    else:
        return "Nötr"

textblob_result = textblob_sentiment_analysis(text)
print("TextBlob Analizi:", textblob_result)
def vader_sentiment_analysis(text):
    analyzer = SentimentIntensityAnalyzer()
    scores = analyzer.polarity_scores(text)
    compound = scores['compound']
    if compound >= 0.05:
        return "Olumlu"
    elif compound <= -0.05:
        return "Olumsuz"
    else:
        return "Nötr"

vader_result = vader_sentiment_analysis(text)
print("VADER Analizi:", vader_result)
textblob_result = textblob_sentiment_analysis(text)
vader_result = vader_sentiment_analysis(text)

print("TextBlob Analizi:", textblob_result)
print("VADER Analizi:", vader_result)
```
